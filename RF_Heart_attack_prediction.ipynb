Data pre-processing

```python
# separate the data and label
X = heart_disease_dataset.drop('Heart Disease', axis=1)
```

Label Binarization

```python
Y = heart_disease_dataset['Heart Disease'].apply(lambda y_value: 1 if y_value>='Presence' else 0)
```

Train & Test Data

```python
X_train, X_test, Y_train, Y_test = train_test_split(X, Y, test_size = 0.2, random_state =3)
```

```python
print(Y.shape, Y_train.shape, Y_test.shape)
```

Random Forest Classifier Model

```python
model = GaussianNB()
```

```python
model.fit(X_train, Y_train)
```

Evaluate Model ||  Accuracy Score

```python
# Accuracy check on test data

X_test_prediction = model.predict(X_test)
test_data_accuracy = accuracy_score(X_test_prediction, Y_test)
```

```python
print(test_data_accuracy * 100)
```

Building a Predictive system 

```python
input_data =(64,1,1,110,211,0,2,144,1,1.8,2,0,3)

# Changing the input data to a numpy array

input_data_as_numpy_array = np.array(input_data)

#Reshape the data as we are predicting the label for only one instance
input_data_reshape = input_data_as_numpy_array.reshape(1, -1)

prediction = model.predict(input_data_reshape)
print(prediction)

if (prediction[0]==1):
  print("Heart disease found.'\n\n")
else:
  print("Heart disease not found.\n\n")
```
