Under-sampling - *Build a Dataset that contain similiar distribution of normal transaction and fraud transaction*


```python
legit_sample = legit.sample(n=492)
```

```python
#Concatenating two DataFrams
new_dataset = pd.concat([legit_sample, fraud], axis = 0)
```

```python
new_dataset.groupby('Class').mean()
```

Spiliting data into feature and terget

```python
X = new_dataset.drop(columns='Class', axis = 1)
Y = new_dataset['Class']
```

Spilting data into traing & testing data

```python
X_train, X_test, Y_train, Y_test = train_test_split( X, Y, test_size=0.2, stratify=Y, random_state=2)
```

Model Training

> Logistics Regression

```python
print(X.shape, X_train.shape, X_test.shape)
```

```python
#model = LogisticRegression()
model = RandomForestClassifier()
# model = KNeighborsClassifier()
#model = DecisionTreeClassifier()
#model = SVC()
```

```python
#train data with logistic regression
model.fit(X_train, Y_train)
```

```python
#accuracy result
X_train_predection = model.predict(X_train)
training_data_accuracy = accuracy_score(X_train_predection, Y_train)
```

```python
print('Accuracy for training data:', training_data_accuracy * 100)
```

```python
X_test_prediction = model.predict(X_test)
testing_data_accuracy = accuracy_score(X_test_prediction, Y_test)
```

```python
print('Accuracy on Test data is:', testing_data_accuracy * 100)
```
